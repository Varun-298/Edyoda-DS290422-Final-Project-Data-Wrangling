```python
# 1. Import the dataset using Pandas from above mentioned url.
import pandas as pd
Covid=pd.read_csv("https://raw.githubusercontent.com/SR1608/Datasets/main/covid-data.csv")
Covid
```

```python
# 2. High Level Data Understanding: 
# a. Find no. of rows & columns in the dataset
Covid.shape
```

```python
# Data types of columns.
Covid.dtypes
```

```python
# c. Info & describe of data in dataframe.
Covid.info()
```

```python
# 3. Low Level Data Understanding : 
# a. Find count of unique values in location column.
count=Covid["location"].nunique()
print(count)
```

```python
# b. Find which continent has maximum frequency using values counts.
max_freq_count=Covid["continent"].value_counts()
print(max_freq_count)
max_freq_count.max
```

```python
# c. Find maximum & mean value in 'total_cases'.
Covid["total_cases"].max()
```

```python
Covid["total_cases"].mean()
```

```python
# d. Find 25%,50% & 75% quartile value in 'total_deaths'.
Covid.total_deaths.quantile([0.25,0.5,0.75])
```

```python
# e. Find which continent has maximum 'human_development_index'.
Covid[["continent","human_development_index"]].max()
```

```python
Covid.groupby("continent").agg({"human_development_index":"max"})
```

```python
# f. Find which continent has minimum 'gdp_per_capita'.
Covid.groupby("continent").agg({"gdp_per_capita":"min"})
```

```python
# 4. Filter the dataframe with only this columns ['continent','location','date','total_cases','total_deaths','gdp_per_capita',' human_development_index'] and update the data frame.
Covid[["continent", "location", "date", "total_cases", "total_deaths", "gdp_per_capita", "human_development_index"]]
```

```python
# 5. Data Cleaning 
# a. Remove all duplicates observations
Covid=Covid.drop_duplicates()
```

```python
# b. Find missing values in all columns
Covid.isna()
```

```python
Covid.isna().sum()
```

```python
# c. Remove all observations where continent column value is missing
Covid=Covid.dropna(subset=["continent"])
Covid["continent"].isna().sum()
```

```python
# d. Fill all missing values with 0
Covid=Covid.fillna(0)
Covid
```

```python
# 6. Date time format : 
# a. Convert date column in datetime format using pandas.to_datetime
Covid["date"]=pd.to_datetime(Covid["date"])
Covid["date"]
```

```python
# b. Create new column month after extracting month data from date column.
Covid["month"]=pd.DatetimeIndex(Covid["date"]).month
Covid["month"]
```

```python
# 7. Data Aggregation: 
# a. Find max value in all columns using groupby function on 'continent' column
Covid.groupby("continent").max().reset_index()
```

```python
Covid
```

```python
# b. Store the result in a new dataframe named 'df_groupby'. (Use df_groupby dataframe for all further analysis)
df_groupby=Covid
df_groupby
```

```python
# 8. Feature Engineering : a. Create a new feature 'total_deaths_to_total_cases' by ratio of 'total_deaths' column to 'total_cases'
df_groupby["total_deaths_to_total_cases"]=df_groupby["total_deaths"]/df_groupby["total_cases"]
df_groupby
```

```python
# 9. Data Visualization : 
# a. Perform Univariate analysis on 'gdp_per_capita' column by plotting histogram using seaborn dist plot.
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.set_style("whitegrid")
sns.distplot(df_groupby["gdp_per_capita"], kde=False, color="blue", bins=40)
```

```python
# b. Plot a scatter plot of 'total_cases' & 'gdp_per_capita'
sns.scatterplot(x="gdp_per_capita",y="total_cases",data=df_groupby)
```

```python
# c. Plot Pairplot on df_groupby dataset.
sns.pairplot(data=df_groupby,hue="continent")
plt.show()
```

```python
# d. Plot a bar plot of 'continent' column with 'total_cases' .
g=sns.barplot(x="continent",
             y="total_cases",
              kind="bar",
             data=df_groupby)
```
